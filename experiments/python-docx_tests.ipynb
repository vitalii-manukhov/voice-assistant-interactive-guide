### Test docx

```python
from docx import Document

filepath = "./data/test_docx.docx"

doc = Document(filepath)

print(doc.tables)
print("Количество таблиц:", len(doc.tables))
print(doc.tables[0])
```

```python
for table_numer in range(len(doc.tables)):
    table = doc.tables[table_numer]
    print("\nТаблица:", table_numer, '\n')
    for row in table.rows:
        for cell in row.cells:
            print(cell.text)
```
